# Import pandas Library

```python
import pandas as pd
df = pd.read_csv("Dummy_Sales_Data_v1.csv")
```

# Print Data Frame


```python
df
```

# Data Frame .Head() Funtion

```python
#df.head(): Function which returns the first 5 rows of the dataset.
#you can access the First Rows according to you Need.
#df.head(10):It returns the first 10 Rows of your data Set
df.head()
```

# Data Frame .Tail() Function

```python
#This function helps you to get the last few rows of the dataset. By default,
#it returns the last5 rows  of df. you can simply mention the desired number of rows
#df.tail(10) it returns the last 10 rows of your data set
df.tail()
```

# Data Frame.Sample() Function

```python
#The sample() method returns a specified number of random rows. 
#The sample() method returns 1 row if a number is not specified.
df.sample(1)
```

# Data Frame().info Function

```python
#This function returns a  information  of the DataFrame.
#This  information aboutcolumn names and their respective data types,missing values.
df.info()
```

# Data Frame.describe() Function

```python
#This function returns  statistics values about the data. This includes
#minimum,maximum, mean and  standard deviation  of the values in columns.
df.describe()
```

# Data Frame .Query() Function

```python
#This function is used to query the DataFrame based on an expression. An expression
#can be as simple as a single condition and as complex as a combination of multiple conditions.
df.query("Quantity < 100")
```

# Data Frame.Loc() Function

```python
#This function  returns the group of rows and columnsidentified by their 
#labels or names.
df.loc[50,'Status']
```

# Data Frame.Unique() Function

```python
df["Product_Code"].unique()
```

# Data Frame.Nunique() Function

```python
#The nunique() method returns the number of unique values for each column.
#By specifying the column axis ( axis='columns' ) For columns.
df.nunique()
```

# Data Frame.is Null() Function

```python
#The isnull() method returns a DataFrame object where all the values are 
#replaced with a Boolean value True for NULL values, and otherwise False
df.isnull()
```

# Data Frame.Fillna() Function

```python
#The fillna() method replaces the NULL values with a specified value. 
df.fillna(222)
```

# Data Frame.Sort_Values() Function

```python
#In order to sort the data frame in pandas, function sort_values() is used.
df.sort_values(by='Quantity')
```

# Data Frame.value_counts() Function

```python
#Return a Series containing counts of unique values.
#The resulting object will be in descending order.
df["Product_Category"].value_counts()
```

# Data Frame.nlargest() Function

```python
#The nlargest() method returns a specified number of rows.
#starting at the top after sorting the DataFrame by the highest value for a specified column.
df.nlargest(5 , "Quantity")
```

# Data Frame.nsmallest() Function

```python
#The nsmallest method is used to return the first n rows in ascending order with the smallest values in columns.
df.nsmallest(5 , "Quantity")
```

# Data Frame.copy() Function

```python
#The copy() method returns a copy of the DataFrame.
#By default, the copy is a "deep copy" meaning that any changes made in the original DataFrame will NOT be reflected in the copy.
df.copy()
```

# Data Frame.rename() Function

```python
#The rename() function is used to alter axes labels.
df.rename(columns= {"Quantity" : "Product Quantity" , "Sales_Manager" : "Manager"})
```

# Data Frame.where()

```python
#DataFrame. where() function is similar to if else that is used to check the 
#one or multiple conditions of an expression in DataFrame and replace with another value when the condition becomes False
condition = df["Quantity"] > 20
result = df.where(condition , other = 1 )
result
```

#  Data Frame.drop() Function

```python
#The drop() method removes the specified row or column.
#By specifying the column axis ( axis='columns' ), the drop() method removes the specified column
df.drop("Quantity", axis='columns')
```

# Data Frame.corr() Function

```python
#corr() method is to find the pairwise correlation of all the columns in the DataFrame. If any null value is present,
#Sit will automatically be excluded.
result = df.corr()
result
```
